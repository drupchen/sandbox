```python
#Now do Sanskrit: Skt.vowels, [g|d|b|dz]+_h, hr, shr, Skt
Hregex1 = r"([ཀ-ཬཱ-྅ྐ-ྼ]{0,}[ཱཱཱིུ-ཹཻཽ-ྃ][ཀ-ཬཱ-྅ྐ-ྼ]{0,}|[ཀ-ཬཱ-྅ྐ-ྼ]{0,}[གཌདབཛྒྜྡྦྫ][ྷ][ཀ-ཬཱ-྅ྐ-ྼ]{0,}|[ཀ-ཬཱ-྅ྐ-ྼ]{0,}[ཤཧ][ྲ][ཀ-ཬཱ-྅ྐ-ྼ]{0,}|[ཀ-ཬཱ-྅ྐ-ྼ]{0,}[གྷཊ-ཎདྷབྷཛྷཥཀྵ-ཬཱཱཱིུ-ཹཻཽ-ྃྒྷྚ-ྞྡྷྦྷྫྷྵྐྵ-ྼ][ཀ-ཬཱ-྅ྐ-ྼ]{0,})"
# more Sanskrit: invalid superscript-subscript pairs
Hregex2 = r"([ཀ-ཬཱ-྅ྐ-ྼ]{0,}[ཀཁགང-ཉཏ-དན-བམ-ཛཝ-ཡཤཧཨ][ྐ-ྫྷྮ-ྰྴ-ྼ][ཀ-ཬཱ-྅ྐ-ྼ]{0,})"
# tsa-phru mark used in Chinese transliteration
Hregex3 = r"([ཀ-ཬཱ-྅ྐ-ྼ]{0,}[༹][ཀ-ཬཱ-྅ྐ-ྼ]{0,})"
```

```python
content = prepareTib(test).syls_only()
```

```python
new = []
for word in content:
    if re.search(Hregex1, word):
        new.append(word+'#1')
    elif re.findall(Hregex2, word):
        new.append(word+'#2')
    elif re.findall(Hregex3, word):
        new.append(word+'#3')
    else:
        new.append(word)
print(new)
```

```python
output = '་\n'.join(new)
filewrite('namjom.txt', output)
```
