```python
import re
import os
```

```python
def isTibetanLetter(c):
    if (c >= 'ༀ' and c <= '༃') or (c >= 'ཀ' and c <= 'ྼ'):
        return True
    return False
```

```python
# find all non-Tibetan characters in a file
punct = ['༄', '༅', '།', '་', '༌', '༑', '༎', '༏', '༐', '༔']
chars = []
for File in os.listdir('./texts'):
    with open('./texts/'+File, 'r', -1, 'utf-8-sig') as f:
        content = [line.strip() for line in f.readlines()]
        for line in content:
            for letter in line:
                if not isTibetanLetter(letter) and letter not in chars and letter not in punct:
                    chars.append(letter)
print(chars)
```

```python
# open file, delete all segmentation characters, deletes spaces except between shad, writes the text as lines not separated by the newlines
# replaces all files in place
for File in os.listdir('./texts'):
    with open('./texts/'+File, 'r', -1, 'utf-8-sig') as f:
        content = f.read()
        content = re.sub(r'\n+', '', content)
    truc = re.sub(r'[\t\(\)\ufeff\[\]\+\-\!ne\xa0\x07\x08]+', '', content) # content of chars
    truc = re.sub(r'([^།༑]) ', r'\1', truc)
    new = [line for line in re.split('[།༑]+', truc) if line != '']
    with open('./texts/'+File, 'w', -1, 'utf-8-sig') as f:
        f.write('\n'.join(new))
```
